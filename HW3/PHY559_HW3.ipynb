### 3. Calculate the deceleration parameter $q_0$ and $H_0t_0$ for models with the following $\Omega$s:

#### (a) $\Omega_{m,0}=2$, $\Omega_{r,0}=0$, $\Omega_{v,0}=0$

#### (b) $\Omega_{m,0}=1$, $\Omega_{r,0}=0$, $\Omega_{v,0}=0$

#### (c) $\Omega_{m,0}=0.25$, $\Omega_{r,0}=0$, $\Omega_{v,0}=0$

#### (d) $\Omega_{m,0}=0$, $\Omega_{r,0}=0$, $\Omega_{v,0}=0$

#### (e) $\Omega_{m,0}=0$, $\Omega_{r,0}=1$, $\Omega_{v,0}=0$

#### (f) $\Omega_{m,0}=0.27$, $\Omega_{r,0}=0$, $\Omega_{v,0}=0.73$

#### (g) $\Omega_{m,0}=1$, $\Omega_{r,0}=0$, $\Omega_{v,0}=1.095$

#### (h) $\Omega_{m,0}=0$, $\Omega_{r,0}=1$, $\Omega_{v,0}=1.595$

```python
'''Required Libraries'''
import matplotlib.pyplot as plt
import numpy as np

'''Required Functions'''
from scipy.integrate import quad
```

```python
'''Function to calculate the scaled age of the universe'''
def age(a):
    H0t0 = 1 / ((1 - m0 + r0 + v0) + (r0 / a**2) + (m0 / a) + (v0 * a**2))
    return H0t0
```

```python
'''Given densities to test'''
omega_m0 = [2, 1, 0.25, 0, 0, 0.27, 1, 0]       #density of matter
omega_r0 = [0, 0, 0, 0, 1, 0, 0, 1]             #density of radiation
omega_v0 = [0, 0, 0, 0, 0, 0.73, 1.095, 1.595]  #density of vacuum


'''Set and populate table'''
table = []
labels = ['\u03A9_m0', '\u03A9_r0', '\u03A9_v0', 'q0', 'H0\u00B7t0']
for m0, r0, v0 in zip(omega_m0, omega_r0, omega_v0):
    
    '''Caluclate q0 and H0t0'''
    q0 = (0.5 * m0) + r0 - v0
    H0t0 = quad(age, 0, 1)[0]
    
    '''Append row to table'''
    table.append([m0, r0, v0, f'{q0:.3f}', f'{H0t0:.3f}'])
    
    
'''Create figure and plot results'''
fig, ax = plt.subplots(1, 1, figsize = (8, 5))
fig.tight_layout(pad = 3)
ax.axis('off')
ax.annotate('Comparison of Relevant Data for Different Models', xy = (0.5, 0.95),
            xycoords = 'axes fraction', ha = 'center', va = 'center', fontsize = 15, fontweight = 'bold')
the_table = ax.table(cellText = table, colLabels = labels, loc = 'center', cellLoc = 'center')
the_table.scale(1, 2)
for (row, col), cell in the_table.get_celld().items():
    if (row == 0) or (col == -1):
        cell.set_text_props(fontweight = 'bold')
        
'''Save figure'''
fig.savefig('c:/Users/Tucker Knaak/Downloads/DensitiesTable.png')
```
